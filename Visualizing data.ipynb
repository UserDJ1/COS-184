Now histogram each column of the data. One way to do this is to loop over all of the columns and use DataFrame's own <code>hist()</code> method, whose first argument is the column number you want to histogram. Notice anything odd about the second and next-to-last columns? You can take a look at the raw data, or wait until we make a boxplot -- that will make everything clear.

Now compute the mean and standard deviation of the data in each column (via the <code>mean()</code> and <code>std()</code> methods of a <code>DataFrame</code>) and print them.

```python
# Your code goes here, to compute a list of the means and standard deviations of each column.

print(
    pd.concat( [data.mean(), data.std()] , axis=1 ).rename( columns={ 0:"Mean" , 1:"Std. Dev." } ))
```

Hmmm. It seems as if the means and standard deviations are strictly increasing.

Now let's make a boxplot of the entire data. The easiest way to do this is to make use of the <code>boxplot()</code> method of <code>DataFrame</code>.

```python
# Don't change this cell.
# Makes a boxplot. Use the following as flier properties.
f = dict(marker='o', markersize=10, markerfacecolor='blue')
_ = data.boxplot(flierprops=f, return_type='dict')
```

Whoa! Looks like the data is nice and consistent EXCEPT for a few "close" outliers and two wild ones! See if you can spot the wild ones in the raw data and leave a comment here.
